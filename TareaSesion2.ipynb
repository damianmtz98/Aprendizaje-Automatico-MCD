```python
# Eliminamos nuestra columna numero 15
df_hpdatos = df_hpdatos.drop(columns = 'Exercise angina' )
df_hpdatos.head(5)
```

```python
# Filtramos la edad que sean mayor a 70 con la frecuencia cardiaca de mayor de 140.
ageheart = df_hpdatos[(df_hpdatos['Age'] > 65 ) & (df_hpdatos['Max HR'] > 100 ) & (df_hpdatos ['BP'] > 130)]
df_ageheart = pd.DataFrame(ageheart)
df_ageheart
```

```python
# Creamos un subconjunto de datos
# Donde tendremos la edad, presion arterial, resultados del electrocardiograma y las enfermedades del corazon.
subcon1 = df_ageheart[['Age', 'BP', 'Max HR', 'Sex']]
df_subcon1 = pd.DataFrame(subcon1)
df_subcon1
```

```python
# Creamos este grupo donde buscamos los hombres que sean mas vulnerables arriba de 65 años.
gpo1 = df_subcon1.groupby(['Age'])[['Sex']].sum()
gpo1 
```
